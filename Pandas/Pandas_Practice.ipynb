# Road runner

```python
df = {
    'distance': {0: 82,   1: 101,   2: 84,   3: 106,   4: 93,   5: 86,   6: 87,   7: 82,   8: 92,   9: 86,   10: 98,   11: 84},
    'runner': {0: 'runner1',   1: 'runner2',   2: 'runner3',   3: 'runner4',   4: 'runner1',   5: 'runner2',   6: 'runner3',   7: 'runner4',   8: 'runner1',   9: 'runner2',   10: 'runner3',   11: 'runner4'}}
df = pd.DataFrame(df)
```

```python
import pandas as pd

def get_count(df):
    '''
    INPUT: df -> dataframe
    
    OUTPUT: result -> series
    '''
    ### CODE STARTS HERE
    def count_of_each(df1):
        return df1["distance"] - df1["distance"].mean()
        
    grouped = df.groupby("runner").apply(count_of_each).reset_index()
    grouped.drop("level_1", axis=1, inplace=True)
    result = grouped.loc[grouped["distance"]>0]
    result = result.groupby('runner').size()    
    return result
```

```python
get_count(df)
```

# Mean Income

```python
df = pd.DataFrame(
    {'name': ['Elon', 'Jeff', 'Bill', 'Falguni',], 
    'gender': ['M', 'F', 'M', 'F',],
    'income': [53000., 28000., 25000., 44000.]}
)
```
