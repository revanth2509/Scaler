```python
def binarysearch(arr, n):
    l = 0
    r = len(arr)
    while l<r:  #n
        mid = (l+r)//2  #n/2
        if arr[mid] == n:
            return mid
        elif arr[mid]<n:
            l = mid+1
        else:
            r = mid-1
```
